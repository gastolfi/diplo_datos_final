## Análisis de outliers

Para visualizar los valores fuera de rango u outliers los identificamos mediante el método de quartiles.

Referencia de experience_level

*   EN - Entry-level
*   MI - Junior Mid-level
*   SE - Intermediate Senior-level
*   EX - Expert Executive-level / Director

```python
# df3_salarios es el DataFrame y queremos analizar los outliers de la columna 'salary_in_usd'

salary_analysis = 'salary_in_usd'

# Calcular los límites del bigote (whiskers) en el boxplot
Q1 = df3_salarios[salary_analysis].quantile(0.25)
Q3 = df3_salarios[salary_analysis].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar los valores atípicos basados en los límites del bigote
outliers = df3_salarios[(df3_salarios[salary_analysis] < lower_bound) | (df3_salarios[salary_analysis] > upper_bound)]

# Imprimir los valores atípicos
outliers_info = outliers[[salary_analysis,'work_year','job_title','experience_level','employment_type']]
```

```python
outliers_info
```

```python
# Filtrar los valores atípicos basados en los valores de 'experience_level'
experience_filter = outliers_info['experience_level'].isin(['EN', 'MI'])
filtered_outliers = outliers_info[experience_filter]
```

```python
# Imprimir los valores atípicos con 'experience_level' igual a 'EN' o 'MI'
filtered_outliers
```

```python
#Eliminación de outliers
df4_salarios = df3_salarios.drop(outliers_info[experience_filter].index)
```

## Eliminación de Freelance y Contract

Eliminamos los datos de freelance y contract porque son solo 10 de cada uno, alrededor del 0.4% de los datos. Lo que es poco para hacer estadistica y también podría llegar a influir en análisis previos por no tener una relación directa del tiempo de trabajo con el sueldo percibido.

```python
salarios_sinflct = df4_salarios[df4_salarios['employment_type'].isin(['FT', 'PT']) & (df4_salarios['employment_type'] != 'FL')& (df4_salarios['employment_type'] != 'CT')]
```

## Columna salario 160hs

```python
# Función para calcular el salario según 160hs de trabajo
def calcular_salarioft(row):
    if row['employment_type'] == 'FT':
        return row['salary_in_usd']
    elif row['employment_type'] == 'PT':
        return row['salary_in_usd'] * 2

# Aplicar la función para crear la nueva columna
salarios_sinflct['salary_in_usd_160hs'] = salarios_sinflct.apply(calcular_salarioft, axis=1)
```
